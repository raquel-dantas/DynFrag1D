## Compare symmetry 

```python
#  Compare symmetry for Lip-field

file_address = 'LOG/test_symmetry/uniform_mesh/lipfield/half_bar_300el/lipfield_' 

u_lip_half = readResults(file_address, "u")
v_lip_half = readResults(file_address, "v")
acel_lip_half =readResults(file_address, "acel")
d_lip_half = readResults(file_address, "d")
n_fragments_lip_half = readResults(file_address, "n_fragments")
avg_frag_size_lip_half = readResults(file_address, "avg_frag_size")
avg_stress_bar_lip_half = readResults(file_address, "avg_stress_bar")
energies_lip_half = readResults(file_address, "energies")
u_all_steps_lip_half = readResults(file_address, "u_all_steps")
damage_all_steps_lip_half = readResults(file_address, "damage_all_steps")
stress_all_steps_lip_half = readResults(file_address, "stress_all_steps")
fraglen_all_steps_lip_half = readResults(file_address, "fraglen_all_steps")
time_data = readResults(file_address, "time_data")
time_simulation = time_data[0]
n_steps = time_data[2]

file_address = 'LOG/test_symmetry/uniform_mesh/lipfield/whole_bar_600el/lipfield_' 
u_lip_whole = readResults(file_address, "u")
v_lip_whole = readResults(file_address, "v")
acel_lip_whole =readResults(file_address, "acel")
d_lip_whole = readResults(file_address, "d")
n_fragments_lip_whole = readResults(file_address, "n_fragments")
avg_frag_size_lip_whole = readResults(file_address, "avg_frag_size")
avg_stress_bar_lip_whole = readResults(file_address, "avg_stress_bar")
energies_lip_whole = readResults(file_address, "energies")
u_all_steps_lip_whole = readResults(file_address, "u_all_steps")
damage_all_steps_lip_whole = readResults(file_address, "damage_all_steps")
stress_all_steps_lip_whole = readResults(file_address, "stress_all_steps")
# fraglen_all_steps_lip_whole = readResults(file_address, "fraglen_all_steps")
```

```python
# Separate the results you want to compare

time = np.linspace(0, time_simulation, n_steps)

stress_bar = [
    ["whole bar", avg_stress_bar_lip_whole],
    ["half bar", avg_stress_bar_lip_half],
]

plotCompareSimulations(
    stress_bar,
    time,
    title="Test use symmetry",
    label_x="time (s)",
    label_y="Average stress at the bar",
    save_filename="LOG/test_symmetry_avg_stress.svg",
)

nfrag = [
    ["whole bar", n_fragments_lip_whole],
    ["half bar", n_fragments_lip_half],
]

plotCompareSimulations(
    nfrag,
    time,
    title="Test use symmetry",
    label_x="time (s)",
    label_y="Number of fragments",
    save_filename="LOG/test_symmetry_nfrag.svg",
)
```

```python
file_address = 'LOG/mesh_study/lipfield_uniform_mesh/input_1/625el/lipfield_' 

d_625el = readResults(file_address, "d")
n_fragments_625el = readResults(file_address, "n_fragments")
avg_stress_bar_625el = readResults(file_address, "avg_stress_bar")
var_energies_625el = readResults(file_address, "var_energies")
time_data = readResults(file_address, "time_data")
time_simulation = time_data[0]
n_steps_625el = time_data[2]


file_address = 'LOG/mesh_study/lipfield_uniform_mesh/input_1/1250el/lipfield_' 

d_1250el = readResults(file_address, "d")
n_fragments_1250el = readResults(file_address, "n_fragments")
avg_stress_bar_1250el = readResults(file_address, "avg_stress_bar")
var_energies_1250el = readResults(file_address, "var_energies")
time_data_1250el = readResults(file_address, "time_data")
n_steps_1250el = time_data_1250el[2]


file_address = 'LOG/mesh_study/lipfield_uniform_mesh/input_1/2500el/lipfield_' 

d_2500el = readResults(file_address, "d")
n_fragments_2500el = readResults(file_address, "n_fragments")
avg_stress_bar_2500el = readResults(file_address, "avg_stress_bar")
var_energies_2500el = readResults(file_address, "var_energies")
time_data_2500el = readResults(file_address, "time_data")
n_steps_2500el = time_data_2500el[2]


file_address = 'LOG/mesh_study/lipfield_uniform_mesh/input_1/5000el/lipfield_' 

d_5000el = readResults(file_address, "d")
n_fragments_5000el = readResults(file_address, "n_fragments")
avg_stress_bar_5000el = readResults(file_address, "avg_stress_bar")
var_energies_5000el = readResults(file_address, "var_energies")
time_data_5000el = readResults(file_address, "time_data")
n_steps_5000el = time_data_5000el[2]




# Separate the results you want to compare

time_625el = np.linspace(0, time_simulation, n_steps_625el)
time_1250el = np.linspace(0, time_simulation, n_steps_1250el)
time_2500el = np.linspace(0, time_simulation, n_steps_2500el)
time_5000el = np.linspace(0, time_simulation, n_steps_5000el)



stress_bar = [
    ["625el", time_625el, avg_stress_bar_625el],
    ["1250el", time_1250el, avg_stress_bar_1250el],
    ["2500el", time_2500el, avg_stress_bar_2500el],
    ["5000el", time_5000el, avg_stress_bar_5000el],
]

plotCompareSimulations(
    stress_bar,
    title="Input1: Average stress",
    label_x="time (s)",
    label_y="Average stress at the bar",
    save_filename="LOG/mesh_study/lipfield_uniform_mesh/input_1/avg_stress.svg",
)


nfrag = [
    ["625el", time_625el, n_fragments_625el],
    ["1250el", time_1250el, n_fragments_1250el],
    ["2500el", time_2500el, n_fragments_2500el],
    ["5000el", time_5000el, n_fragments_5000el],
]

plotCompareSimulations(
    nfrag,
    title="Input1: Number of fragments",
    label_x="time (s)",
    label_y="Number of fragments",
    save_filename="LOG/mesh_study/lipfield_uniform_mesh/input_1/number_fragments.svg",
)

# Damage

bar_length = 0.5 * 5. * 10**-3
x0 = 0.
xf = bar_length

node_coord_625 = np.linspace(x0, xf, 625 + 1)
node_coord_1250 = np.linspace(x0, xf, 1250 + 1)
node_coord_2500 = np.linspace(x0, xf, 2500 + 1)
node_coord_5000 = np.linspace(x0, xf, 5000 + 1)

x_625el = np.array([[node_coord_625[el], node_coord_625[el + 1]]for el in range(625)])
x_625el = x_625el.flatten()*1000
x_1250el = np.array([[node_coord_1250[el], node_coord_1250[el + 1]]for el in range(1250)])
x_1250el = x_1250el.flatten()*1000
x_2500el = np.array([[node_coord_2500[el], node_coord_2500[el + 1]]for el in range(2500)])
x_2500el = x_2500el.flatten()*1000
x_5000el = np.array([[node_coord_5000[el], node_coord_5000[el + 1]]for el in range(5000 )])
x_5000el = x_5000el.flatten()*1000


d_625el = np.array([[d_625el[el], d_625el[el]] for el in range(625)])
d_625el = d_625el.flatten()
d_1250el = np.array([[d_1250el[el], d_1250el[el]] for el in range(1250)])
d_1250el = d_1250el.flatten()
d_2500el = np.array([[d_2500el[el], d_2500el[el]] for el in range(2500)])
d_2500el = d_2500el.flatten()
d_5000el = np.array([[d_5000el[el], d_5000el[el]] for el in range(5000)])
d_5000el = d_5000el.flatten()


d = [
    ["625el", x_625el, d_625el],
    ["1250el", x_1250el, d_1250el],
    ["2500el", x_2500el, d_2500el],
    ["5000el", x_5000el, d_5000el],
]
plotCompareSimulations(
    d,
    title="Input1: d",
    label_x="x (mm)",
    label_y="Damage",
    save_filename="LOG/mesh_study/lipfield_uniform_mesh/input_1/damage.svg",
)

# Compare smallest and biggest mesh
d = [
    ["625el", x_625el, d_625el],
    ["5000el", x_5000el, d_5000el],
]
plotCompareSimulations(
    d,
    title="Input1: d",
    label_x="x (mm)",
    label_y="Damage",
    save_filename="LOG/mesh_study/lipfield_uniform_mesh/input_1/damage_2.svg",
)


# Convergence number fragments
meshes = [625, 1250, 2500, 5000]

# Convergence dissipated energy
plotConvergence(
    nfrag,
    meshes, 
    title="Input1: number fragments convergence",
    label_x="mesh",
    label_y="Number of fragments",
    save_filename="LOG/mesh_study/lipfield_uniform_mesh/input_1/number_fragments_convergence.svg",
)


```
