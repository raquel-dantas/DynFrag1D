## LIP: uniform mesh

```python
filepath = "LOG/mesh_study/lipfield_uniform_mesh/"

file_address = filepath + "2500el/lipfield_"
u_2500el = readResults(file_address, "u")
v_2500el = readResults(file_address, "v")
acel_2500el =readResults(file_address, "acel")
d_2500el = readResults(file_address, "d")
n_fragments_2500el = readResults(file_address, "n_fragments")
avg_frag_size_2500el = readResults(file_address, "avg_frag_size")
avg_stress_bar_2500el = readResults(file_address, "avg_stress_bar")
energies_2500el = readResults(file_address, "energies")
u_all_steps_2500el = readResults(file_address, "u_all_steps")
damage_all_steps_2500el = readResults(file_address, "damage_all_steps")
stress_all_steps_2500el = readResults(file_address, "stress_all_steps")
fraglen_all_steps_2500el = readResults(file_address, "fraglen_all_steps")
time_data_2500el = readResults(file_address, "time_data")
n_steps_2500el = time_data_2500el[2]
time_simulation = time_data_2500el[0]


file_address = filepath + "5000el/lipfield_"
d_5000el = readResults(file_address, "d")
n_fragments_5000el = readResults(file_address, "n_fragments")
avg_frag_size_5000el = readResults(file_address, "avg_frag_size")
avg_stress_bar_5000el = readResults(file_address, "avg_stress_bar")
time_data_5000el = readResults(file_address, "time_data")
n_steps_5000el = time_data_5000el[2]


file_address = filepath + "7500el/lipfield_"
d_7500el = readResults(file_address, "d")
n_fragments_7500el = readResults(file_address, "n_fragments")
avg_frag_size_7500el = readResults(file_address, "avg_frag_size")
avg_stress_bar_7500el = readResults(file_address, "avg_stress_bar")
time_data_7500el = readResults(file_address, "time_data")
n_steps_7500el = time_data_7500el[2]
```

```python
# Separate the results you want to compare

time_2500el = np.linspace(0, time_simulation, n_steps_2500el)
time_5000el = np.linspace(0, time_simulation, n_steps_5000el)
time_7500el = np.linspace(0, time_simulation, n_steps_7500el)



stress_bar = [
    ["2500el", time_2500el, avg_stress_bar_2500el],
    ["5000el", time_5000el, avg_stress_bar_5000el],
    ["7500el", time_7500el, avg_stress_bar_7500el],
]

plotCompareSimulations(
    stress_bar,
    title="Uniform mesh LIP: Average stress",
    label_x="time (s)",
    label_y="Average stress at the bar",
    save_filename="LOG/mesh_study/lipfield_uniform_mesh/avg_stress.svg",
)


nfrag = [
    ["2500el", time_2500el, n_fragments_2500el],
    ["5000el", time_5000el, n_fragments_5000el],
    ["7500el", time_7500el, n_fragments_7500el],
]

plotCompareSimulations(
    nfrag,
    title="Uniform mesh LIP: Number of fragments",
    label_x="time (s)",
    label_y="Number of fragments",
    save_filename="LOG/mesh_study/lipfield_uniform_mesh/number_fragments.svg",
)


# Convergence number fragments
meshes = [2500, 5000, 7500]

# Convergence dissipated energy
plotConvergence(
    nfrag,
    meshes, 
    title="Uniform mesh LIP: number fragments convergence",
    label_x="mesh",
    label_y="Number of fragments",
    save_filename="LOG/mesh_study/lipfield_uniform_mesh/number_fragments_convergence.svg",
)


```
