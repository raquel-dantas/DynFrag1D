```python
import numpy as np

import DFProcessResults
import DFPlotResults
```

```python
# This script is for a mesh study: compare several meshes
# Initialize what you want to compare:
avg_stress_10to3_uniform_mesh = []
n_fragments_10to3_uniform_mesh  = []
avg_stress_10to3_non_uniform_mesh  = []
n_fragments_10to3_non_uniform_mesh = []
```

```python
#  Path to the output files
filepath_10to3 = "output_mesh_study/10to3/"
```

```python
# 10to3: Lip-field 625 elements Uniform mesh
simulation_name = "LIP 625 elements"
file_address = filepath_10to3 + "10to3_lipfield_uniform_625/lipfield_"
time_data = DFProcessResults.getTimeData(file_address)
time_simulation = time_data[0]
dt = time_data[1]
n_steps = time_data[2]
n_files = int(n_steps / 10 + 1)
avg_stress_bar, energies, n_fragments = DFProcessResults.getResultsAllSteps(file_address, n_files, n_steps)
var_energies = DFProcessResults.computeVarEnergiesLipfield(energies, n_files)
time = np.linspace(0, time_simulation, n_files)
avg_stress_10to3_uniform_mesh.append([simulation_name, time, avg_stress_bar])
n_fragments_10to3_uniform_mesh.append([simulation_name, time, n_fragments])

DFPlotResults.plotResults(
    avg_stress_bar * 1e-6,
    time,
    label_x="time (s)",
    label_y="Average stress at the bar (MPa)",
    plot_title="10to3 LIP 625 uniform elements: Average stress at the bar",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/10to3_lipfield_uniform_625_avgstress"
)
DFPlotResults.plotResults(
    n_fragments,
    time,
    label_x="time (s)",
    label_y="N",
    plot_title="10to3 LIP 625 uniform elements: Number of fragments",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/10to3_lipfield_uniform_625_nfrag"
)

DFPlotResults.plotVarEnergiesLipfield(
    var_energies,
    time,
    plot_title="10to3 625 uniform elements LIP: Variation of energy",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/10to3_lipfield_uniform_625_energy"
)
```
