```python
from matplotlib import pyplot as plt

# meshes = [625, 1250, 2500, 5000, 7500]
meshes = [1250, 2500, 5000, 7500]
lip_nfrag_uniform = [
    # ["625 elements", time_uniform_lip625, n_fragments_uniform_lip625[2]],
    ["1250 elements", time_uniform_lip1250, n_fragments_uniform_lip1250[2]],
    ["2500 elements", time_uniform_lip2500, n_fragments_uniform_lip2500[2]],
    ["5000 elements", time_uniform_lip5000, n_fragments_uniform_lip5000[2]],
    ["7500 elements", time_uniform_lip7500, n_fragments_uniform_lip7500[2]],
]
lip_nfrag_non_uniform = [
    # ["625 elements", time_non_uniform_lip625, n_fragments_non_uniform_lip625[2]],
    ["1250 elements", time_non_uniform_lip1250, n_fragments_non_uniform_lip1250[2]],
    ["2500 elements", time_non_uniform_lip2500, n_fragments_non_uniform_lip2500[2]],
    ["5000 elements", time_non_uniform_lip5000, n_fragments_non_uniform_lip5000[2]],
    ["7500 elements", time_non_uniform_lip7500, n_fragments_non_uniform_lip7500[2]],
]
czm_nfrag_uniform = [
    # ["625 elements", time_uniform_czm625, n_fragments_uniform_czm625[2]],
    ["1250 elements", time_uniform_czm1250, n_fragments_uniform_czm1250[2]],
    ["2500 elements", time_uniform_czm2500, n_fragments_uniform_czm2500[2]],
    ["5000 elements", time_uniform_czm5000, n_fragments_uniform_czm5000[2]],
    ["7500 elements", time_uniform_czm7500, n_fragments_uniform_czm7500[2]],
]
czm_nfrag_non_uniform = [
    # ["625 elements", time_non_uniform_czm625, n_fragments_non_uniform_czm625[2]],
    ["1250 elements", time_non_uniform_czm1250, n_fragments_non_uniform_czm1250[2]],
    ["2500 elements", time_non_uniform_czm2500, n_fragments_non_uniform_czm2500[2]],
    ["5000 elements", time_non_uniform_czm5000, n_fragments_non_uniform_czm5000[2]],
    ["7500 elements", time_non_uniform_czm7500, n_fragments_non_uniform_czm7500[2]],
]

lip_ediss_uniform = [
    # ["625 elements", time_uniform_lip625, energy_diss_uniform_lip625[2]],
    ["1250 elements", time_uniform_lip1250, energy_diss_uniform_lip1250[2]],
    ["2500 elements", time_uniform_lip2500, energy_diss_uniform_lip2500[2]],
    ["5000 elements", time_uniform_lip5000, energy_diss_uniform_lip5000[2]],
    ["7500 elements", time_uniform_lip7500, energy_diss_uniform_lip7500[2]],
]
lip_ediss_non_uniform = [
    # ["625 elements", time_non_uniform_lip625, energy_diss_non_uniform_lip625[2]],
    ["1250 elements", time_non_uniform_lip1250, energy_diss_non_uniform_lip1250[2]],
    ["2500 elements", time_non_uniform_lip2500, energy_diss_non_uniform_lip2500[2]],
    ["5000 elements", time_non_uniform_lip5000, energy_diss_non_uniform_lip5000[2]],
    ["7500 elements", time_non_uniform_lip7500, energy_diss_non_uniform_lip7500[2]],
]
czm_ediss_uniform = [
    # ["625 elements", time_uniform_czm625, energy_diss_uniform_czm625[2]],
    ["1250 elements", time_uniform_czm1250, energy_diss_uniform_czm1250[2]],
    ["2500 elements", time_uniform_czm2500, energy_diss_uniform_czm2500[2]],
    ["5000 elements", time_uniform_czm5000, energy_diss_uniform_czm5000[2]],
    ["7500 elements", time_uniform_czm7500, energy_diss_uniform_czm7500[2]],
]
czm_ediss_non_uniform = [
    # ["625 elements", time_non_uniform_czm625, energy_diss_non_uniform_czm625[2]],
    ["1250 elements", time_non_uniform_czm1250, energy_diss_non_uniform_czm1250[2]],
    ["2500 elements", time_non_uniform_czm2500, energy_diss_non_uniform_czm2500[2]],
    ["5000 elements", time_non_uniform_czm5000, energy_diss_non_uniform_czm5000[2]],
    ["7500 elements", time_non_uniform_czm7500, energy_diss_non_uniform_czm7500[2]],
]






def plotConvergenceCZMLIP(results0, results1, results2, results3,  meshes, title, label_x, label_y, save_filename):

    fig, axes = plt.subplots()
    axes.grid(True, which="both")
    axes.axhline(y=0, color="k")
    plt.title(title)
    plt.xlabel(label_x)
    plt.ylabel(label_y)

    nb_simulations = len(meshes)
    x_values = np.zeros(nb_simulations)
    final_values0 = np.zeros(nb_simulations)
    final_values1 = np.zeros(nb_simulations)
    final_values2 = np.zeros(nb_simulations)
    final_values3 = np.zeros(nb_simulations)

    for i in range(nb_simulations):
        x_values[i] = meshes[i]
        final_values0[i] = max(results0[i][2])
        final_values1[i] = max(results1[i][2])
        final_values2[i] = max(results2[i][2])
        final_values3[i] = max(results3[i][2])
    plt.plot(x_values, final_values0, marker='.',label='LIP: Uniform mesh', color='steelblue')
    plt.plot(x_values, final_values1, marker='.',label="LIP: Non-uniform mesh", color='steelblue', linestyle='dotted')
    plt.plot(x_values, final_values2, marker='.',label="CZM: Uniform mesh", color='darkorange')
    plt.plot(x_values, final_values3, marker='.',label="CZM: Non-uniform mesh", color='darkorange', linestyle='dotted')

    plt.legend()
    plt.savefig(save_filename)
    plt.show()


plotConvergenceCZMLIP(
    lip_nfrag_uniform, 
    lip_nfrag_non_uniform, 
    czm_nfrag_uniform, 
    czm_nfrag_non_uniform, 
    meshes, 
    title="10to5: number fragments convergence", 
    label_x="Number of elements",
    label_y="Number of fragments", 
    save_filename="output_mesh_study/plots_mesh_study/10to5_convergence_nfrag.svg")
plotConvergenceCZMLIP(
    lip_ediss_uniform, 
    lip_ediss_non_uniform, 
    czm_ediss_uniform, 
    czm_ediss_non_uniform, 
    meshes, 
    title="10to5: energy dissipation convergence", 
    label_x="Number of elements",
    label_y="Dissipated Energy ($ KJ/ {m^2} $)", 
    save_filename="output_mesh_study/plots_mesh_study/10to5_convergence_ediss.svg")
```
