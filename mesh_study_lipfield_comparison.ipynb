```python
import numpy as np

import DFProcessResults
import DFPlotResults
```

```python
# This script is for a mesh study: compare several meshes
# Initialize what you want to compare:
avg_stress_10to3_uniform_mesh = []
avg_stress_10to4_uniform_mesh = []
avg_stress_10to5_uniform_mesh = []


n_fragments_10to3_uniform_mesh  = []
n_fragments_10to4_uniform_mesh  = []
n_fragments_10to5_uniform_mesh  = []

energy_diss_10to3_uniform_mesh = []
energy_diss_10to4_uniform_mesh = []
energy_diss_10to5_uniform_mesh = []


avg_stress_10to3_non_uniform_mesh = []
avg_stress_10to4_non_uniform_mesh = []
avg_stress_10to5_non_uniform_mesh = []


n_fragments_10to3_non_uniform_mesh  = []
n_fragments_10to4_non_uniform_mesh  = []
n_fragments_10to5_non_uniform_mesh  = []

energy_diss_10to3_non_uniform_mesh = []
energy_diss_10to4_non_uniform_mesh = []
energy_diss_10to5_non_uniform_mesh = []

```

```python
#  Path to the output files
filepath_10to3 = "output_mesh_study/10to3/"
filepath_10to4 = "output_mesh_study/10to4/"
filepath_10to5 = "output_mesh_study/10to5/"
```

```python
# 10to4: Lip-field 1250 elements Uniform mesh
simulation_name = "LIP 1250 elements"
file_address = filepath_10to4 + "10to4_lipfield_uniform_1250/lipfield_"
n_elements = 1250
time_data = DFProcessResults.getTimeData(file_address)
time_simulation = time_data[0]
dt = time_data[1]
n_steps = time_data[2]
n_files = int(n_steps / 10 + 1)
avg_stress_bar, energies, n_fragments = DFProcessResults.getResultsAllSteps(file_address, n_files, n_steps)
energy_diss = DFProcessResults.getDissipatedEnergyLipfield(energies, n_files)
var_energies = DFProcessResults.computeVarEnergiesLipfield(energies, n_files)
time = np.linspace(0, time_simulation, n_files)

avg_stress_10to4_uniform_mesh.append([simulation_name, time, avg_stress_bar])
n_fragments_10to4_uniform_mesh.append([simulation_name, time, n_fragments])
energy_diss_10to4_uniform_mesh.append([simulation_name, n_elements, energy_diss])
```
