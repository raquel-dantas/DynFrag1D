```python
import numpy as np

import DFProcessResults
import DFPlotResults
```

```python
# This results are related to the regularization length study
bar_length = 50e-3 #(m)
n_elements = 5000
h = bar_length / n_elements * 1e3 #(mm)
#  Path to the output files
filepath = "output_mesh_study/reg_length/"
```

```python
# l = 1.25 elements
simulation_name = "1.25 elements"
file_address = filepath + "5000_10to4_uniform_reg_length_1andquarter/lipfield_"
reg_length_1andquarter = 1.25 * h
time_data = DFProcessResults.getTimeData(file_address)
time_simulation = time_data[0]
dt = time_data[1]
n_steps = time_data[2]
n_files = int(n_steps / 10 + 1)
avg_stress_bar, energies, n_fragments = DFProcessResults.getResultsAllSteps(
    file_address, n_files, n_steps
)
energy_diss = (
    DFProcessResults.getDissipatedEnergyLipfield(energies, n_files) * 1e-3
)  # from J to KJ
var_energies_1andquarter = DFProcessResults.computeVarEnergiesLipfield(
    energies, n_files
)
time = np.linspace(0, time_simulation, n_files)

avg_stress_1andquarter = [simulation_name, time, avg_stress_bar]
n_fragments_1andquarter = [simulation_name, time, n_fragments]
energy_diss_1andquarter = [simulation_name, reg_length_1andquarter, energy_diss]
```

```python
DFPlotResults.plotVarEnergiesLipfield(
    var_energies_1andquarter,
    time,
    plot_title="Variation of energy (l = 1.25 h)",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/energy_1andquarter",
)

final_step = n_steps - n_steps % 10
damage_1andquarter = DFProcessResults.getResultFinalStep(
    "damage", file_address, final_step
)
DFPlotResults.plotDamage(
    damage_1andquarter,
    n_elements,
    label_x="x (mm)",
    label_y="Damage",
    plot_title="Damage (l = 1.25 h)",
    save_plot=False,
    save_filename="s",
)
```

```python
avg_stress_comparison = []
n_fragments_comparison  = []
energy_diss_comparison = []
reg_lengths = []

reg_lengths.append(reg_length_1andquarter)
reg_lengths.append(reg_length_2andhalf)
reg_lengths.append(reg_length_5)
reg_lengths.append(reg_length_10)

avg_stress_comparison.append(avg_stress_1andquarter)
avg_stress_comparison.append(avg_stress_2andhalf)
avg_stress_comparison.append(avg_stress_5)
avg_stress_comparison.append(avg_stress_10)

n_fragments_comparison.append(n_fragments_1andquarter)
n_fragments_comparison.append(n_fragments_2andhalf)
n_fragments_comparison.append(n_fragments_5)
n_fragments_comparison.append(n_fragments_10)


energy_diss_comparison.append(energy_diss_1andquarter)
energy_diss_comparison.append(energy_diss_2andhalf)
energy_diss_comparison.append(energy_diss_5)
energy_diss_comparison.append(energy_diss_10)
```

```python
DFPlotResults.plotResultsComparison(
    avg_stress_comparison,
    label_x="time (s)",
    label_y="Average stress at the bar (MPa)",
    plot_title="Average stress at the bar",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/reg_length_comp_avgstress"
)
DFPlotResults.plotResultsComparison(
    n_fragments_comparison,
    label_x="time (s)",
    label_y="N",
    plot_title="Number of fragments",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/reg_length_comp_nfrag"
)

DFPlotResults.plotConvergence(
    energy_diss_comparison,
    reg_lengths,
    label_x="regularizarion length (mm)",
    label_y="Dissipated Energy ($ J/ {m^2} $)",
    plot_title="Dissipated energy",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/reg_length_convergence_ediss"
)

DFPlotResults.plotConvergence(
    n_fragments_comparison,
    reg_lengths,
    label_x="regularizarion length (mm)",
    label_y="N",
    plot_title="Number of fragments",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/reg_length_convergence_nfrag"
)
```
