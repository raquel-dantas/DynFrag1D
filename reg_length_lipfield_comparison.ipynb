```python
DFPlotResults.plotResultsComparison(
    avg_stress_comparison,
    label_x="time (s)",
    label_y="Average stress at the bar (MPa)",
    plot_title="Average stress at the bar",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/reg_length_comp_avgstress"
)
DFPlotResults.plotResultsComparison(
    n_fragments_comparison,
    label_x="time (s)",
    label_y="N",
    plot_title="Number of fragments",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/reg_length_comp_nfrag"
)


reg_lengths = [1.25 * 50e-3 / 5000, 2.5 * 50e-3 / 5000, 5 * 50e-3 / 5000, 10 * 50e-3 / 5000] 

# DFPlotResults.plotConvergence(
#     energy_diss_comparison,
#     reg_lengths,
#     label_y="Dissipated Energy ($ kJ/ {m^2} $)",
#     plot_title="10to4 LIP uniform: Dissipated energy",
#     save_plot=True,
#     save_filename="output_mesh_study/plots_mesh_study/reg_length_convergence_ediss"
# )

# DFPlotResults.plotConvergence(
#     n_fragments_comparison,
#     reg_lengths,
#     label_y="N",
#     plot_title="Number of fragments",
#     save_plot=True,
#     save_filename="output_mesh_study/plots_mesh_study/reg_length_convergence_nfrag"
# )

from matplotlib import pyplot as plt

def plotConvergence(
    results: list,
    x: list,
    label_x: str,
    label_y: str,
    plot_title: str,
    save_plot: bool,
    save_filename: str,
):
    fig, axes = plt.subplots()
    axes.grid(True, which="both")
    axes.axhline(y=0, color="k")
    plt.title(plot_title)
    plt.xlabel(label_x)
    plt.ylabel(label_y)

    nb_simulations = len(results)
    x_values = np.zeros(nb_simulations)
    final_values = np.zeros(nb_simulations)

    for i in range(nb_simulations):
        x_values[i] = x[i]
        final_values[i] = max(results[i][2])
    plt.plot(x_values, final_values, marker=".")

    plt.legend()
    if save_plot == True:
        plt.savefig(save_filename + ".svg")
    plt.show()


plotConvergence(
    energy_diss_comparison,
    reg_lengths,
    label_x="regularizarion length (m)",
    label_y="Dissipated Energy ($ kJ/ {m^2} $)",
    plot_title="10to4 LIP uniform: Dissipated energy",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/reg_length_convergence_ediss"
)

plotConvergence(
    n_fragments_comparison,
    reg_lengths,
    label_x="regularizarion length (m)",
    label_y="N",
    plot_title="Number of fragments",
    save_plot=True,
    save_filename="output_mesh_study/plots_mesh_study/reg_length_convergence_nfrag"
)
```
