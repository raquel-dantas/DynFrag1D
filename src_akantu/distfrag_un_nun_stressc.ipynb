```python
def PlotFragmentSizeHistogram(frag_sizes):

    fig, axes = plt.subplots()
    axes.grid(True, which='both')
    axes.axhline(y=0, color='k')
    plt.title("Fragment size distribution")
    plt.xlabel("Fragment size (mm)")
    plt.ylabel("Number of fragments")

    # lim_min = min(frag_sizes)
    # lim_max = max(frag_sizes)
    plt.hist(frag_sizes*1000,5)
    # plt.hist(frag_sizes_un, 10)
    # plt.hist(frag_sizes_nun, 10)

    # plt.savefig("LOG/fragment_size_distribution.svg")
    plt.show()






PlotFragmentSizeHistogram(fraglen_un)
PlotFragmentSizeHistogram(fraglen_nun)
# PlotFragmentSizeHistogram(fraglen_un, fraglen_nun)





```

```python
def PlotFragmentSizeHistogram(frag_sizes_un, frag_sizes_nun):

    hfont = {'fontname':'sans-serif'}
    fig, axes = plt.subplots()
    axes.grid(True, which='both')
    axes.axhline(y=0, color='k')
    # plt.title("Fragment size distribution", **hfont)
    plt.xlabel("Fragment size (mm)",  **hfont)
    plt.ylabel("Number of fragments",  **hfont)
    plt.rcParams.update({'font.size': 12})

    # lim_min = min(frag_sizes)
    # lim_max = max(frag_sizes)
    # plt.hist(frag_sizes*1000,5)
    plt.hist(frag_sizes_un*1000, 20, histtype='stepfilled', alpha=0.4,  label='Uniform critical stress', )
    plt.hist(frag_sizes_nun*1000, 20, histtype='stepfilled', alpha=0.4,  label='Non-Uniform critical stress')
    plt.legend()
    plt.xlim(0,2.25, 0.25)
    plt.savefig("fragment_size_distribution.pdf")
    # plt.show()






# PlotFragmentSizeHistogram(fraglen_un)
# PlotFragmentSizeHistogram(fraglen_nun)
PlotFragmentSizeHistogram(fraglen_un, fraglen_nun)
```
